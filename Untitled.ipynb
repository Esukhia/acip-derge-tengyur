```python
import requests
import re
from bs4 import BeautifulSoup
```

```python
text_id = "D1793"
web_text = requests.get(f"https://www.istb.univie.ac.at/kanjur/rktsneu/etanjur/verif2.php?id={text_id}&coll=derge")
```

```python
soup = BeautifulSoup(web_text.text)
```

```python
pages = soup.findAll("font", {"size" : "+2"})
```

```python
page_ann = re.findall("<hr>.+?<br>", web_text.text)
```
